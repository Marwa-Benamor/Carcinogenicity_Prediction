```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Descriptors

# Initialiser une liste pour stocker les descripteurs calculÃ©s pour chaque molÃ©cule
descriptors_list = []

# Loop Ã  travers les SMILES et calculer les descripteurs pour chaque molÃ©cule
for smiles in smiles_list:
    try:
        mol = Chem.MolFromSmiles(smiles)
        # Si la molÃ©cule est invalide, passer Ã  la suivante
        if mol is None:
            continue
        # Si la molÃ©cule est valide, calculer les descripteurs
        else:
            mw = Descriptors.MolWt(mol)
            logp = Descriptors.MolLogP(mol)
            hba = Descriptors.NumHAcceptors(mol)
            hbd = Descriptors.NumHDonors(mol)
            tpsa = Descriptors.TPSA(mol)
            num_double_bonds = len([bond for bond in mol.GetBonds() if bond.GetBondType() == Chem.rdchem.BondType.DOUBLE])
            num_nitro_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('[N+](=O)[O-]')))
            num_amine_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('[NH2]')))
            num_hydroxyl_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('[OH]')))
            num_sulfonate_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('S(=O)(=O)[O-]')))
            num_ester_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('C(=O)O')))
            num_ketone_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('C(=O)')))
            num_aldehyde_groups = len(mol.GetSubstructMatches(Chem.MolFromSmarts('C=O')))
            charge = Chem.GetFormalCharge(mol)
            # Ajouter les descripteurs Ã  la liste
            descriptors_list.append([smiles, mw, logp, hba, hbd, tpsa, num_double_bonds,
                                     num_nitro_groups, num_amine_groups, num_hydroxyl_groups,
                                     num_sulfonate_groups, num_ester_groups, num_ketone_groups,
                                     num_aldehyde_groups,charge])
    except:
        # Si la syntaxe SMILES est invalide, passer Ã  la molÃ©cule suivante
        continue

# CrÃ©er un DataFrame Ã  partir de la liste de descripteurs
descriptors_df = pd.DataFrame(descriptors_list, columns=['SMILES', 'MW', 'LogP', 'HBA', 'HBD', 'TPSA',
                                                         'NumDoubleBonds', 'NumNitroGroups', 'NumAmineGroups',
                                                         'NumHydroxylGroups', 'NumSulfonateGroups', 'NumEsterGroups',
                                                         'NumKetoneGroups', 'NumAldehydeGroups','charge'])

# Afficher le DataFrame
print(descriptors_df)
```
